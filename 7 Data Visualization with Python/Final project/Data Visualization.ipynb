```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
survey = {'Very interested': [1332,1688,429,1340,1263,1629], 'Somewhat interested': [729,444,1081,734,770,477], 'Not interested': [127,60,610,102,136,74]}
index1 = ['Big Data (Spark / Hadoop)', 'Data Analysis / Statistics', 'Data Journalism', 'Data Visualization', 'Deep Learning', 'Machine Learning']

df=pd.DataFrame(survey, columns = ['Very interested', 'Somewhat interested', 'Not interested'], index = index1)

df
```

```python
df.sort_values('Very interested', ascending = False, inplace = True)
df
```

```python
df['Very interested_pect'] = round(df['Very interested'] / 22.33, 2).astype(str) + '%'

df['Somewhat interested_pect'] = round(df['Somewhat interested'] / 22.33, 2).astype(str) + '%'
df['Not interested_pect'] = round(df['Not interested'] / 22.33, 2).astype(str) + '%'
df
```

```python
df.index
```

```python
ax =df[['Very interested', 'Somewhat interested', 'Not interested']].plot(kind = 'bar', 
                                                                      width = 0.8, 
                                                                      color=('#5cb85c', '#5bc0de', '#d9534f'), 
                                                                      figsize = (20,8), 
                                                                      fontsize = 14)

plt.title('Percentage of Respondents\'Interest in Data Science Areas', fontsize = 16)

p=0
for n in df.index:
    i = 0
    for y in ['Very interested', 'Somewhat interested', 'Not interested']:
        l=[-0.35,-0.1,0.2]
        q = l[i]
        plt.annotate(df.loc[n, y+'_pect'],
                xy = (p+q, df.loc[n,y]),
                va = 'bottom',
                ha = 'left')
        i = i+1
    p = p+1
    
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.yaxis.set_visible(False)
    
plt.show()
```
